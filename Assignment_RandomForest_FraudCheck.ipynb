**Use Random Forest to prepare a model on fraud data, treating those who have taxable_income <= 30000 as "Risky" and others are "Good"**

```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
fraud_df= pd.read_csv('/content/Fraud_check.csv')
```

```python
fraud_df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
fraud_df['Undergrad']=label_encoder.fit_transform(fraud_df['Undergrad'])
fraud_df['Marital.Status']= label_encoder.fit_transform(fraud_df['Marital.Status'])
fraud_df['Urban'] = label_encoder.fit_transform(fraud_df['Urban'])
```

```python
fraud_df
```

```python
tax_val=[]
for value in fraud_df['Taxable.Income']:
  if value > 30000:
    tax_val.append('Good')
  else :
      tax_val.append('Risky')
fraud_df['tax_val']= tax_val
```
